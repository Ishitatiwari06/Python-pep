```python
# prime
n=int(input())
fact=2
isprime=True
for i in range(2,n):
  if(n%i==0):
    isprime=False
    break;
if(isprime):
  print("Prime")
else:
  print("Not Prime")
```

```python
# food delivery
from datetime import date

users={}
user=input()
if user not in users.keys():
  users[user]=date.today()
  print("Congrats you will be getting 70% discount")
else:
  if((users[user]-date.today())<7):
    print("Will get 50% off")
  elif((users[user]-date.today())<20):
    print("will get 25% off")
  else:
    print("no discount for you")
```

```python
# login
users={"Ishita":"1234","Shreya":"shreya123"}
name=input()
password=input()
if name in users.keys():
  if users[name]==password:
    print("Logged in successfully")
  else:
    print("Wrong password")
else:
  print("User does not exists")
```

```python
# atm machine
users={"1234567890":["8080",40000],"9876543210":["9906",3000]}
card_no=input("Enter card number:")
if(card_no in users.keys()):
  pin=input("Enter pin(4 digit):")
  if(len(pin)!=4):
    print("Pin should be exactly 4 digits long")
  else:
    if(users[card_no][0]==pin):
      amt=int(input("Enter amount:"))
      if(amt<=10000 and users[card_no][1]>=amt):
        print("allowed to proceed")
      elif(users[card_no][1]<amt):
        print("Insufficient balance")
      else:
        print("Amount should not exceed 10000")
else:
  print("Invalid")

        


      

```
