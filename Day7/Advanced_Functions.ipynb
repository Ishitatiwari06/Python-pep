```python
# Scope
city = "Delhi"
def func():
  global city
  print(city)
print(city)
func()
```

```python
balance=30
def deposit(n):
  global balance
  balance+=n
def withdraw(n):
  global balance
  balance-=n
```

```python
# sum
sum=lambda x,y:x+y
print(sum(8,9))
# multiply
multiply=lambda x,y:x*y
print(multiply(4,5))
```

```python
# largest
a=int(input("Enter a number 1:"))
b=int(input("Enter a number 2:"))
res=lambda x,y:max(x,y)
print(res(a,b))
```

```python
# prime or not
from functools import reduce
n=int(input("Enter a number:"))
prime=lambda ans:"Prime" if reduce(lambda x,y:x and y,map(lambda i:n%i!=0,range(2,n))) else "Not Prime"
print(prime(n))
```

```python
# Celcius to Fahrenheit
temp=int(input("Enter a number:"))
ctof=lambda x:x*(9/5)+32
print(ctof(temp))
```

```python
# Multiplication Table
n=int(input("Enter number:"))
table=list(map(lambda i:n*i,range(1,11)))
print(table)
```

```python
# Cube of 5 nos
l=[1,2,3,4,5]
cubes=list(map(lambda x:x**3,l))
print(cubes)
```

```python
# even/odd
n=int(input("Enter number"))
even_odd=lambda x:"Even" if n%2==0 else "Odd"
print(even_odd(n))
```

```python
# positive/negative
n=int(input("Enter number"))
pos_neg=lambda x:"Positive" if x>0 else "Negative" if x<0 else "Zero"
print(pos_neg(n))
```

```python
# Apply discount of 20% on every product
l=[30,25,50,40,89]
dist=list(map(lambda x:x-x*0.2,l))
print(dist)
```

```python
# increment n to [n+1,n+2...n+20]
n=int(input("Enter number:"))
inc=list(map(lambda x:x+n,range(1,21)))
print(inc)
```

```python
# convert name into uppercase
name=input("Enter name:")
uppercase=lambda x:name.upper()
print(uppercase(name))
```

```python
# check fees >100000 in array
l=[10000,300000,230988,450000,5500000]
fee=list(filter(lambda x:x>10000,l))
print(fee)
```
