file.readline() - only one line
file.readlines() - read all lines as a list
file.tell() - gives the current cursor position
file.seek() - used to change cursor position
x - mode to create a file

```python
file=open("employees.txt","x")
file.write("Hello this is my first file\n")
file.close()
```

```python
with open("employees.txt","r") as file:
  print(file.read())
  file.close()
```

```python
# no of words
with open("MyFile","r") as file:
  line=file.read()
print(len(line.split(" ")))
file.close()
```

```python
# check if word exists
word=input()
file=open("MyFile","r")
line=file.read()
words=(line.split(" "))
if word in words:
  print(word,"exists in your file")
else:
  print(word,"does not exists in your file")
file.close()
```

```python
# copy to another file
file2=open("MyCopy.txt","w")
file=open("MyFile",'r')
content=file.read()
file2.write(content)
file.close()
file2.close()
```

```python
# convert to uppercase
file=open("MyFile","r")
line=file.read()
file.close()
file=open("MyFile","w")
file.write(line.upper())
file.close()
```
