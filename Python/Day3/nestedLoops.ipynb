```python
for i in range(3):
  for j in range(3):
    print(i,j)
  
```

```python
for i in range(1,4):
  print("Row",i)
  for j in range(1,4):
    print(" Column",j)
```

```python
for i in range(3):
  for j in range(4):
    print("*",end=" ")
  print()
```

```python
for i in range(3):
  for j in range(1,5):
    print(j,end=" ")
  print()
```

```python
for i in range(3):
  for j in range(4):
    print(chr(65+j),end=" ")
  print()
```

```python
# multiplication grid
for i in range(1,4):
  for j in range(1,4):
    print(i*j,end=" ")
  print()
```

```python
for i in range(1,5):
  for j in range(4):
    print(i,end=" ")
  print()
```

```python
n=1
for i in range(3):
  for j in range(3):
    print(n,end=" ")
    n+=1
  print()
```
