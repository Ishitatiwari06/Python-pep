```python
# abstraction
from abc import ABC,abstractmethod
class Vehicle:
  @abstractmethod
  def fueltype():
    pass
class PetrolCar(Vehicle):
  def fueltype(self):
    print("Petrol")
class ElectricCar(Vehicle):
  def fueltype(self):
    print("Electric")

p=PetrolCar()
e=ElectricCar()
p.fueltype()
e.fueltype()
```

```python
# inheritance
class Person:
  def __init__(self,name,age):
    self._name=name
    self._age=age
class Teacher(Person):
  def __init__(self, name, age,subject):
    super().__init__(name, age)
    self._subject=subject
  def role(self):
    print("Teacher")
class Student(Person):
  def __init__(self, name, age,roll_no):
    super().__init__(name, age)
    self.roll_no=roll_no
  def role(self):
    print("Student")
t=Teacher("Maya",28,"English")
s=Student("Anjali",20,1)
s.role()
t.role()
```

```python
# Polymorphism
class Square:
  def draw(self):
    print("Square")
    for i in range(3):
      print("* * *")
class Triangle:
  def draw(self):
    print("Triangle")
    for i in range(3):
      print("*"*(i+1))
shapes=[Square(),Triangle()]
for shape in shapes:
  shape.draw()
```

```python
# Smart Library System
class Library:
  def __init__(self):
    self.available=True
  def issue(self):
    if(self.available):
      self.available=False
      print("Book Issued")
    else:
      print("Book not available")
  def return_book(self,days):
      self.available=True
      if(days>5):
        print("Late Fees:",10*(days-5))
      else:
        print("Returned book, No late fee applied")
stud=Library()
stud.issue()
stud.return_book(3)
stud.issue()
stud.return_book(6)
```

```python
# Airline Ticket Reservation
class TicketBooking:
  def __init__(self,name,weight,cls):
    self.name=name
    self.weight=weight
    self.cls=cls
  def billing(self):
    if(self.cls=="Economy"):
      self.fare=150*(self.weight)*15
    elif(self.cls=="Business"):
      self.fare=1500*(self.weight)*15
    else:
      self.fare=15000*(self.weight)*15
    print("Boarding Details:")
    print("Name:",self.name)
    print("Baggage Weight:",self.weight)
    print("Total Fare:",self.fare)
book=TicketBooking("Reeta",34,"Economy")
book.billing()
    
```

```python
# Restaurant Billing
class Order:
  gst=53
  def __init__(self,amt):
    self.amt=amt
class Dinein(Order):
  def __init__(self, amt):
    super().__init__(amt)
    self.discount=10
  def billing(self):
    bill=self.amt-(self.amt*self.discount/100)+self.gst
    print("DineIn")
    print("Total amount:",bill)
class takeaway(Order):
  def __init__(self, amt):
    super().__init__(amt)
    self.discount=20
  def billing(self):
    bill=self.amt-(self.amt*self.discount/100)+self.gst
    print("TakeAway")
    print("Total amount:",bill)

order=Dinein(3000)
order.billing()
print(order.gst)
```
