```python
rows=[['Aditi',21],['Rohan',23],['Ravi'],[None,34],[],['Aditi',31]]
df=pd.DataFrame(rows,columns=['Name','Age'])
print(df)
```

```python
df.info()
df.describe()
df.columns
df.index
df['Name'].value_counts() #freq of each unique value
```

```python
df[df['Name'].isin(['Aditi','Raghav'])]
df.query("Age>22 and Name=='Ishita'")
df.isnull().sum()
df.notnull() #opposite of isnull
```

```python
df.dropna(subset=['Age']) #drops only where age is missing
df.dropna(how='all') #drops only rows where every col is null
```

```python
df['Gender']=['Female','Male','Male','Female',None]
df['Gender']=df['Gender'].replace({'Male':'M','Female':'F'})
df=df.rename(columns={'Name':'Stud_Name'})
df
```

```python
df.duplicated() #check if col have duplicates and return bool value
df.drop_duplicates()
df.drop_duplicates(subset=['Name'],keep='last') #keep- first,last
```

```python
df['Eligible']=df['Age'].map(lambda e: 'Adult' if e>=18 else "Child" if e<=10 else "Teenager")
df
```
