```python
# Student Marks Analysis System
import numpy as np
stud=np.array(['Aditi','Saurab',"Gauri","Chaitanya","Ravi","Guru"])
sub=np.array(["English","Science","Maths","Computer","Hindi"])
marks=np.array([[89,98,90,32,54],[54,65,67,87,98],[74,78,53,85,87],[75,86,32,75,86],[43,75,87,97,34],[64,76,86,43,76]])
avg_marks=np.mean(marks,axis=0)
print(avg_marks)
highest=np.max(marks,axis=0)
lowest=np.min(marks,axis=0)
print(highest)
print(lowest)
std=np.std(marks,axis=0)
print(std)

# compute per-student averages for ranking and top-students
stud_avg = np.mean(marks, axis=1) #row wise

# rank students by their average (highest first)
sort_idx = np.argsort(stud_avg)[::-1]
rank = stud[sort_idx]
print(rank)

total_avg=np.mean(marks)
print(total_avg)
# select students whose average is above overall average
top_stud = stud[stud_avg > total_avg]
print(top_stud)

```

```python
# subject wise analysis

subject_stats = []
for i, subject in enumerate(sub):
    scores = marks[:, i] #all rows but col i
    avg_score = np.mean(scores)
    max_score = np.max(scores)
    min_score = np.min(scores)
    std_score = np.std(scores)
    top_student = stud[np.argmax(scores)]
    low_student = stud[np.argmin(scores)]

    subject_stats.append((subject, avg_score, max_score, top_student, min_score, low_student, std_score))

for subject, avg_score, max_score, top_student, min_score, low_student, std_score in subject_stats:
    print(f"{subject}: Avg = {avg_score:.2f}, Highest = {max_score} ({top_student}), Lowest = {min_score} ({low_student}), Std = {std_score:.2f}")
```

```python
# percentage calculation
tot_marks=np.sum(marks,axis=1)
per=(tot_marks/500)*100
per
```

```python
import matplotlib.pyplot as plt
# plot per-student averages (stud has 6 entries, use stud_avg which matches)
plt.xlabel("Student Names")
plt.ylabel("Avg Marks")
plt.bar(stud, stud_avg,)
plt.show()
```

```python
# subject wise max-min score
import pandas as pd

df_sub = pd.DataFrame({'Highest': highest, 'Lowest': lowest}, index=sub)

df_sub.plot(kind='bar')
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Subject-wise Highest and Lowest Scores')
plt.show()
```
