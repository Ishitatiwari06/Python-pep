remove vs discard
remove - gives error if value is not found
discard - does nothing if not found
pop() - pop out random elem since set is unordered
add() - to add elems to set

set operations-
| - union
& - intersection
- - difference
^ - symmetric difference (all the elements from each set except the common elems)

```python
s1={1,2,3,4}
s2={3,4,5,6}
print(s1.union(s2))
# or
print(s1 | s2)
print(s1.intersection(s2))
# or
print(s1 & s2)
print(s1.difference(s2))
# or
print(s1 - s2)
print(s1.symmetric_difference(s2))
# or
print(s1^s2)
```
