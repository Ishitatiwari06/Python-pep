```python
# Student Performance Project

import pandas as pd
# load dataset
df=pd.read_csv(r"C:\Users\ishit\Downloads\student_performance_project.csv")
df
```

```python
# look at data quality
df.describe()
df.info()
```

```python
# data cleaning
df['Math']=df['Math'].fillna(df['Math'].mean())
df['Science']=df['Science'].fillna(df['Science'].mean())
df['English']=df['English'].fillna(df['English'].mean())
df.info()
```

```python
# removing duplicates
print(df.shape)
df=df.drop_duplicates()
print(df.shape)
```

```python
# filtering students
res = df[(df['Math'] > 50) & (df['Science'] > 50) & (df['English'] > 50)]
res

df['Result'] = ((df['Math'] > 50) & (df['Science'] > 50) & (df['English'] > 50)).map({True: 'Pass', False: 'Fail'})
df[['Math', 'Science', 'English', 'Result']].head()
```

```python
# Student - wise Analysis
df['total_marks'] = df[['Math', 'Science', 'English']].sum(axis=1)
df[['Math', 'Science', 'English', 'total_marks']].head()
```

```python
# total, avg and ranking
student_df = df.copy()
student_df['Average_Marks'] = student_df[['Math', 'Science', 'English']].mean(axis=1)
student_df['Rank'] = student_df['Average_Marks'].rank(method='average', ascending=False)
student_df = student_df.sort_values(by='Rank')
student_df.head()
```

```python
# Gender wise and city wise analysis
print(df.groupby(['City','Gender'])['Math'].agg(['count','min','max']).round(2))
print(df.groupby(['City','Gender'])['Science'].agg(['count','min','max']).round(2))
print(df.groupby(['City','Gender'])['English'].agg(['count','min','max']).round(2))

```

```python
# combined groupby report
combined_report = df.groupby(['City', 'Gender']).agg(
	students=('Math', 'count'),
	math_min=('Math', 'min'),
	math_max=('Math', 'max'),
	math_mean=('Math', 'mean'),
	science_min=('Science', 'min'),
	science_max=('Science', 'max'),
	science_mean=('Science', 'mean'),
	english_min=('English', 'min'),
	english_max=('English', 'max'),
	english_mean=('English', 'mean')
).round(2)
combined_report
```
