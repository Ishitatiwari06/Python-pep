```python
# find freq of each character in a string
s=input()
freq={}
for i in s:
  if(i in freq):
    freq[i]+=1
  else:
    freq[i]=1
print(freq)
```

```python
# convert string into camelCase
s=input()
words=s.split('_')
camelcase=words[0]
for i in words[1:]:
  camelcase+=i.capitalize()
print(camelcase)
```

```python
# convert name into title case
name=input()
title=name.title()
print(title)
```

```python
# extract domain name from email address
email=input()
domain=email.split('@')[-1]
print(domain)
```

```python
# count number of digits, spaces and special characters
s=input()
dig=0,sp=0,spChars=0
for i in s:
  if(i.isdigit()):
    dig+=1
  elif i==' ':
    sp+=1
  elif not i.isalpha():
    spChars+=1
print("Digits:",dig)
print("Spaces:",sp)
print("Special Characters:",dig)
```

```python
# check if 2 strings are anagrams
s1=input("Enter string 1:")
s2=input("Enter string 2:")
if sorted(s1)==sorted(s2):
  print("Anagrams")
else:
  print("Not anagrams")
```

```python
# find longest word in a sentence
s=input()
idx=0
ans=""
for i in s.split():
  if idx<len(i):
    idx=len(i)
    ans=i

print("Longest word:",ans)
```

```python
# replace every space with _
s=input()
print(s.replace(' ','_'))
```

```python
# extract the file extension from file name
s=input()
ext=s.split('.')[-1]
print(ext)
```

```python
# no. of occurences of word python in a string
s=input()
words=s.split(' ')
cnt=0
for i in words:
  if(i.lower()=="python"):
    cnt+=1
print(cnt)
```

```python
# remove leading and trailing spaces from a string
s=input()
print(s.strip())
```
