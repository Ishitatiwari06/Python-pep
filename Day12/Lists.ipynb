```python
# Debugging Challenge
# copy by reference
a=[1,2,3,4]
b=a #aliasing- 2 or more variables point to the same memory address
b[0]=4
a.append(5)
print("a",a)
print("b",b)

# copy by value - copy.deepcopy(<iterable>) from copy module
import copy
c=copy.deepcopy(a)
c.append(2)
print("c",c)
print("a",a)
```
