```python
# calculator
def calculator(op,a,b=0):
  match op:
    case '+':
      return a+b
    case '-':
      return a-b
    case '*':
      return a*b
    case '/':
      return a/b
    case '//':
      return a//b
    case '%':
      return a%b
    case 'sq':
      return a**2
    case _:
      return a**3
operator=input("Enter operator(+,-,*,/,//,%,sq,cube):")
if(operator not in ['+','-','*','/','//','%','sq','cube']):
  print("Invalid operator")
elif(operator in ['sq','cube']):
  num=int(input("Enter a number:"))
  print(f"{operator}({num})= {calculator(operator,num)}")
else:
  num1=int(input("Enter number 1:"))
  num2=int(input("Enter number 2:"))
  print(f"{num1}{operator}{num2}={calculator(operator,num1,num2)}")
```

types of Arguments-
1. Positional Arguments -- pass the arguments in the same order as parameters are defined
2. Keyword Arguments -- arguments as key:val pair so that order does not matter
3. Default Arguments -- value passed to parameters while function def
4. Arbitrary positional Arguments(*args) -- allows to pass variable number of arguments. 
                                 These arguments are stored in tuple
5. Arbitrary Keyword Arguments(**kwargs) -- variable number of arguments can be passed in key=val format


```python
# Examples of Arguments
# 1 Positional
def demo(name,marks):
  print(name,"scored",marks,"marks")
demo("Ravi",45)
# 2 keyword
demo(marks=50,name="Priya")
# 3 default
def demo(name,marks=0):
  print(name,"scored",marks,"marks")
demo("Ganesh")
# 4 arbitrary
def demo(*args):
  for i in args:
    print(i)
demo(*("Ishita",123))
# 5 variable keyword arguments
def demo(**kwargs):
  for key,val in kwargs.items():
    print(key,val)
demo(name="Shaurya",marks=30,age=13)
```

```python
# final bill of supermarket
# discount 30%, if price>10000 then 50%
def finalPrice(purchase,isdiscount):
  amt=0
  for key,val in purchase:
    amt+=(key*val)
  if(isdiscount=='y'):
    if(amt>10000):
      amt-=(amt*0.5)
    else:
      amt-=(amt*0.3)
  return amt
buy='y'
purchase={}
while(buy=='y'):
  price=float(input("Enter price:"))
  quant=float(input("Enter quantity:"))
  purchase[price]=quant
  buy=input("Want to buy more(y/n):")
dist=input("Discount allowed(y/n):")
print("Final Price",finalPrice(purchase,dist))

```

```python
# pass or fail on basis of marks
# avg should be >45
def result(name,marks):
  s+=sum(marks)
  avg=s/len(marks)
  if(avg>45):
    print(name,"is Passed")
  else:
    print(name,"is Failed")
name=input()
marks=[]
for i in range(5):
  marks.append(input("Enter marks:"))
result(name,marks)
```
