```python
# multiplication table
n=int(input())
for i in range(1,11):
  print(n,"x",i,"=",n*i)
```

```python
# sum of n
n=int(input())
sum=0
for i in range(1,n+1):
  sum+=i
print(sum)
```

```python
# factorial
n=int(input())
fact=1
for i in range(1,n+1):
  fact*=i
print("Factorial of",n,"=",fact)
```

```python
# count digits
n=int(input())
cnt=0
a=n
while(a!=0):
  cnt+=1
  a//=10
print("Number of digits in",n,"=",cnt)
```

```python
# sum of digits
n=int(input())
sum=0
a=n
while(a!=0):
  sum+=a%10
  a//=10
print("Sum of digits of",n,"=",sum)
```

```python
# check number is palindrome
n=int(input())
rev=0
a=n
while(a!=0):
  rem=a%10
  rev=rev*10+rem
  a//=10
if(rev==n):
  print(n,"is a palindrome")
else:
  print(n,"is not a palindrome")
```

```python
# prime
n=int(input())
fact=2
isprime=True
for i in range(2,n):
  if(n%i==0):
    isprime=False
    break;
if(isprime):
  print("Prime")
else:
  print("Not Prime")
```

```python
# fibonacci series
n=int(input())
a=0
b=1
if(n>=1):
  print(a)
if(n>=2):
  print(b)
n-=2
while(n>0):
  c=a+b
  print(c)
  a=b
  b=c
  n-=1
```
