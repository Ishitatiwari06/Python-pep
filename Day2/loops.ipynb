```python
# count digits
n=int(input())
cnt=0
a=n
while(a!=0):
  cnt+=1
  a//=10
print("Number of digits in",n,"=",cnt)
```

```python
# sum of digits
n=int(input())
sum=0
a=n
while(a!=0):
  sum+=a%10
  a//=10
print("Sum of digits of",n,"=",sum)
```

```python
# check number is palindrome
n=int(input())
rev=0
a=n
while(a!=0):
  rem=a%10
  rev=rev*10+rem
  a//=10
if(rev==n):
  print(n,"is a palindrome")
else:
  print(n,"is not a palindrome")
```

```python
# prime
n=int(input())
fact=2
isprime=True
for i in range(2,n):
  if(n%i==0):
    isprime=False
    break;
if(isprime):
  print("Prime")
else:
  print("Not Prime")
```

```python
# fibonacci series
n=int(input())
a=0
b=1
if(n>=1):
  print(a)
if(n>=2):
  print(b)
n-=2
while(n>0):
  c=a+b
  print(c)
  a=b
  b=c
  n-=1
```

```python
# reverse a number
n=int(input("Enter number:"))
print("Original Number:",n)
rev=0
while(n>0):
  rem=n%10
  rev=rev*10+rem
  n//=10
print("Reversed number:",rev)
```
