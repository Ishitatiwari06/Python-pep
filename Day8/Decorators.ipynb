```python
# decorators
# are the functions that take another func as an input and extends or modifies its behavior and returns a new func without changing original func
```

```python
def decorator(func):
  def wrapper(name):
    print("hello")
    func(name)
    print("bye")
  return wrapper #returns func obj

# @decorator
# def greet(name):
#   print(name)
# greet("Ishita")
# or
def greet(name):
  print(name)
# sayHello=decorator(greet)
# sayHello("Ishita")
decorator(greet)("Ishita")
```
