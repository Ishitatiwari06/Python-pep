```python
# Recursion- function that call itself, it must have a base case else it causes RecursionError (stack overflow)
# disadv - more memory,slower due to function calls,stack overflow for deep recusion tree 
```

```python
# factorial
def fact(n):
  if(n==1):
    return 1
  return n*fact(n-1)
print(fact(5))
```

```python
# fibonacci
# 0 based
def fib(n):
  if(n<=1):
    return n
  return fib(n-1)+fib(n-2)
print(fib(5))
```

```python
# sum of first n nos
def calcSum(n):
  if(n==1):
    return 1
  return n+calcSum(n-1)
calcSum(5)
```
